```python
import numpy as np
import pandas as pd
import re

filepath="已清洗招聘数据(2).csv"
df=pd.read_csv(filepath,header=0,encoding="gbk")
```

```python
import re
add = df['城市']
sly= df['薪酬']
edu = df['学历要求']
exp = df['工作经验']
address =[]
salary = []
education = []
experience = []
for i in range(0,len(df)):
    try:
        a = add[i].split('-')
        address.append(a[0])
        #print(address[i])
        s = re.findall(r'\d*\.?\d+',sly[i])
        s1= float(s[0])
        s2 =float(s[1])
        salary.append([s1,s2])
        #print(salary[i])
        education.append(edu[i])
        #print(education[i])
        experience.append(exp[i])
        print(experience[i])
    except:
        pass
```
