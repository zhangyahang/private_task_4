```python
import numpy as np
import pandas as pd
import re
 
filepath="已清洗招聘数据(2).csv"
df=pd.read_csv(filepath,header=0,encoding="gbk")
```

```python
df
```

```python
import re
add = df['城市']
sly= df['薪酬']
edu = df['学历要求']
exp = df['工作经验']
address =[]
salary = []
education = []
experience = []
for i in range(0,len(df)):
    try:
        a = add[i].split('-')
        address.append(a[0])
        #print(address[i])
        s = re.findall(r'\d*\.?\d+',sly[i])
        s1= float(s[0])
        s2 =float(s[1])
        salary.append([s1,s2])
        #print(salary[i])
        education.append(edu[i])
        #print(education[i])
        experience.append(exp[i])
        print(experience[i])
    except:
        pass

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
exp_dic={}#得到工作经验的字典
for x in experience:
    exp_dic[x]=exp_dic.get(x,0)+1
```

```python
edu_dic={}#得到学历要求的字典
for x in education:
    edu_dic[x]=edu_dic.get(x,0)+1
```

```python
add_dic={}
for x in address:
    add_dic[x]=add_dic.get(x,0)+1
```

```python
exp_1=list(exp_dic.keys())
exp_2=list(exp_dic.values())
exp_1=exp_1[:14]+exp_1[15:16]
exp_2=exp_2[:14]+exp_2[15:16]
```

```python
edu_1=list(edu_dic.keys())
edu_2=list(edu_dic.values())
edu_1=edu_1[:12]+edu_1[16:17]
edu_2=edu_2[:12]+edu_2[16:17]
```

```python
add_1=list(add_dic.keys())
add_2=list(add_dic.values())
```

```python
plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['font.serif'] = ['KaiTi']
plt.figure(figsize=(10,5))
plt.title('学历要求的折线图',fontsize=20)
plt.xlabel(u'学历',fontsize=14)#设置x轴，并设定字号大小
plt.ylabel(u'数量',fontsize=14)#设置y轴，并设定字号大小
 
#color：颜色，linewidth：线宽，linestyle：线条类型，label：图例，marker：数据点的类型
plt.plot(edu_1,edu_2,color="deeppink",linewidth=2,linestyle=':',label='数量', marker='o')

 
plt.legend(loc=2)#图例展示位置，数字代表第几象限
plt.savefig('学历要求折线图.png')
plt.show()#显示图像
```

```python

matplotlib.rcParams['font.family']='SimHei'
explode=np.zeros((len(edu_1)))
explode[4]=0.1
plt.figure(figsize=(20, 6.5))
patches,l_text,p_text=plt.pie(edu_2,explode=explode,labels=edu_1,autopct='%.2f%%')
plt.legend(exp_1)
plt.legend(loc='center right')
plt.title('学历要求的饼状图')
plt.axis('equal')

# 设置饼图内文字大小

for t in p_text:
    t.set_size(10)

for t in l_text:
    t.set_size(9)
plt.savefig('学历要求的饼状图.png')
plt.show()
```
