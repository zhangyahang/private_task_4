```python
import numpy as np
import pandas as pd
import re
 
filepath="已清洗招聘数据(2).csv"
df=pd.read_csv(filepath,header=0,encoding="gbk")
```

```python
df
```

```python
b3 =u'万/年'

b5=u'k·12薪'
b6=u'万/月'
li3 = df['薪酬']
#注释部分的print都是为了调试用的
for i in range(0,len(li3)):
    try:
        if b3 in li3[i]:
            x = re.findall(r'\d*\.?\d+',li3[i])
            
            min_ = format(float(x[0])/12*10,'.2f')              #转换成浮点型并保留两位小数
        
            max_ = format(float(x[1])/12*10,'.2f')
            li3[i][1] = min_+'-'+max_+u'千/月'
            print(li3[i][1])
            
            
        if b6 in li3[i]:
            x = re.findall(r'\d*\.?\d+',li3[i])
            #print(x)
            #input()
            min_ = format(float(x[0])*10,'.2f')
            max_ = format(float(x[1])*10,'.2f')
            li3[i][1] = str(min_+'-'+max_+u'千/月')        
        if b5 in li3[i]:
            x = re.findall(r'\d*\.?\d+',li3[i])
            #print(x)
            #input()
            min_ = format(float(x[0]),'.2f')
            max_ = format(float(x[1]),'.2f')
            li3[i][1] = str(min_+'-'+max_+u'千/月')
      
    except:
        pass
```
